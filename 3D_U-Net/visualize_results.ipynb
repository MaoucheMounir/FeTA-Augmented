```python
import matplotlib.pyplot as plt
import re
import numpy as np
import pandas as pd
import os
```

```python
def extract_scores(file_path):
    dice_scores = []
    precision_scores = []
    recall_scores = []
    ventricles_volumes = []
    ventricles_errors = []

    # Expression régulière pour extraire chaque type de valeur
    dice_pattern = r"Dice score:\s([0-9.]+)"
    precision_pattern = r"Precision score:\s([0-9.]+)"
    recall_pattern = r"Recall score:\s([0-9.]+)"
    volume_pattern = r"Ventricles volume:\s([0-9]+)"
    error_pattern = r"Ventricles error:\s([-0-9.]+)"

    with open(file_path, 'r') as file:
        content = file.read()

        # Extraction des valeurs avec findall
        dice_scores = [float(x) for x in re.findall(dice_pattern, content)]
        precision_scores = [float(x) for x in re.findall(precision_pattern, content)]
        recall_scores = [float(x) for x in re.findall(recall_pattern, content)]
        ventricles_volumes = [int(x) for x in re.findall(volume_pattern, content)]
        ventricles_errors = [float(x) for x in re.findall(error_pattern, content)]

    return dice_scores, precision_scores, recall_scores, ventricles_volumes, ventricles_errors

def plot_bar(data, title, ylabel, indices, xlabel="ID patient"):

    plt.figure(figsize=(8, 5))

    plt.bar(indices, data)

    plt.title(title)

    plt.xlabel(xlabel)

    plt.ylabel(ylabel)

    plt.xticks(indices)

    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.savefig(os.path.join("graphea",title+".png"))
    plt.show()



transformation_titles = ["none", "gamma+noise", "gamma", "anisotropy", "biasField", "noise", "anisotropy+noise"]
transformation_titles_long = ["no transform", "random gamma+noise", "random gamma", "random anisotropy", "random bias field", "random noise", "random anisotropy+noise"]
```
